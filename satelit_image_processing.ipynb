**Himawari-8 IR Enhanced**


Pada produk Himawari-8 EH menunjukkan suhu puncak awan yang didapat dari pengamatan radiasi pada panjang gelombang 10.4 mikrometer yang kemudian diklasifikasi dengan pewarnaan tertentu, dimana warna hitam atau biru menunjukkan tidak terdapat pembentukan awan yang banyak (cerah), sedangkan semakin dingin suhu puncak awan, dimana warna mendekati jingga hingga merah, menunjukan pertumbuhan awan yang signifikan dan berpotensi terbentuknya awan Cumulonimbus


Dalam folder image terdapat gambar hasil tangkapan dari satelit Himawari-8 IR mulai dari tanggal 14-18 Oktober 2021

## Blue mask / blue layer of satellite image

Contoh mencari blue layer dalam gambar tangkapan satelit pada tanggal 16 Oktober 2021

```python
# open image 
img = Image.open('/content/Image-Processing/image/16 okt 2021.jpg')
pic = np.array(img)

# show the blue layer
blue_mask = pic[ :, :, 2] < 200
pic[blue_mask] = 0
plt.figure(figsize=(14,14))
plt.imshow(pic)
```

## Red mask / Red layer of satellite image


Contoh mencari red layer dalam gambar tangkapan satelit pada tanggal 16 Oktober 2021

```python
# open image 
img = Image.open('/content/Image-Processing/image/16 okt 2021.jpg')
pic = np.array(img)

# show the red layer
red_mask = pic[ :, :, 0] < 200
pic[red_mask] = 0
plt.figure(figsize=(14,14))
plt.imshow(pic)
```

## Green mask / Green layer of satellite image


Contoh mencari green layer dalam gambar tangkapan satelit pada tanggal 16 Oktober 2021

```python
# open image 
img = Image.open('/content/Image-Processing/image/16 okt 2021.jpg')
pic = np.array(img)

# show the green layer
green_mask = pic[ :, :, 1] < 200
pic[green_mask] = 0
plt.figure(figsize=(14,14))
plt.imshow(pic)
```
